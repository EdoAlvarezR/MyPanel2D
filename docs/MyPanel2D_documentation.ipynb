## Geometry

A panel is define as a vector $P$ between two points $A$ and $B$ pointing from $A$ to $B$; *i.e.,* $\vec{P} = \vec{B}-\vec{A}$, $A$ being $j$ and $B$ being $j+i$ in the figure below. The tangent $\hat{t}$ of this panel is defined as $\hat{t}=\frac{\vec{P}}{|\vec{P}|}$ ($\hat{x}^*$ in figure below), and the normal vector $\hat{n}$ (or $\hat{y}^*$) is defined as the cross product of $\hat{t}$ with an implicit **$\hat{z}$ axis going out the page** in the figure below, *i.e.,* $\hat{n} = \hat{z}\times\hat{t} = -t_2\hat{\imath}+t_1\hat{\jmath}=\hat{y}^*$.

NOTE: Defining $\hat{z}$ as coming out of the page implies that a *clockwise circulation will have to be negative*.

NOTE2: Wit this definition of tangency, it is convenient to define the points of an airfoil in the clockwise direction (*i.e.*, starting at the TE, going from the bottom to top back to the TE).

### Constant Vortex

Defining a constant source strength density $\gamma$ along an individual panel, the induced velocities $u^*$ and $v^*$ (normalized by $\gamma$) in axes $x^*$ and $y^*$ are respectively

\begin{align}
    \bullet \quad & u^*_{i,j} = -\frac{\beta_{i,j}}{2\pi}
    \\
    \bullet \quad & v^*_{i,j} = -\frac{1}{2\pi}\ln{\frac{r_{i,j+1}}{r_{i,j}}}
\end{align}

## Solvers

### Constant Source + Uniform Vortex

This solver defines each panel of the body as both a source and vortex panel, where the source strength density $q$ varies from panel to panel but is constant along each panel, meanwhile the vortex strength density is uniform for all panels. This formulation effectively separates camber (and lift generation) from thickness, as vortex and source.

Control points at each panel will be defined as $CP_i = (A_i+B_i)/2$, then the source strength density and a uniform vortex strength density will be calculated by imposing the no flow condition on a given inflow $V_\infty(\vec{x}, t)$ as explained below.

At the $i$-th control point $\vec{c} = CP_i$, the no flow condition is satisfied iff

\begin{align}
    \hat{n}\cdot(\sum\limits_j q_j G_j^s(\vec{c}) + 
                        \gamma\sum\limits_j G_j^v(\vec{c})) 
            = -\hat{n}\cdot V_{\infty} (\vec{c})
,\end{align}

where $q_j$ is the source strength density of the $j$-th panel, $\gamma$ is the uniform vortex strength density along the entire body, $G_j^s$ and $G_j^v$ are respectively the geometrical factor of source and vortex potential in the velocity induced by the $j$-th panel, and $\hat{n}$ is the outside normal of the $i$-th panel associated to this control point. This equation has N unknown $q_j$ terms associated to the $N$ panels in the body plus the uniform $\gamma$, hence, we have $N+1$ unknowns. $N$ equations like the one above can be formulated corresponding to each of the $N$ control points. In order to close the problem, the Kutta condition is imposed at the top and bottom panels at the trailing edge such as to have the flow reconnecting and leaving the surface of the body at the trailing edge, *i.e.,* 

\begin{align}
    \vec{V}(\vec{c}_1) = \vec{V}(\vec{c}_N)
,\end{align}

with $\vec{V}$ the total velocity resulting from panel-induced velocity plus freestream. This is equivalent to 

\begin{align}
    \sum\limits_j q_j G_j^s(\vec{c}_1) + 
                        \gamma\sum\limits_j G_j^v(\vec{c}_1) + V_\infty(\vec{c}_1)
            = \sum\limits_j q_j G_j^s(\vec{c}_N) + 
                        \gamma\sum\limits_j G_j^v(\vec{c}_N) + V_\infty(\vec{c}_N)
.\end{align}



With the Kutta condition $\vec{V}(\vec{c}_1) = \vec{V}(\vec{c}_N)$ in addition to all $N$ no-flow conditions, we end up with the following closed linear system of equations:

\begin{align}
    \pmatrix{\hat{n}_1\cdot G_1^s(c_1) & \cdots & \hat{n}_1\cdot G_N^s(c_1)
                & \hat{n}_1\cdot \sum\limits_j G_j^v(\vec{c}_1)
             \\
             \vdots
             \\
             \hat{n}_N\cdot G_1^s(c_N) & \cdots & \hat{n}_N\cdot G_N^s(c_N)
                & \hat{n}_N\cdot \sum\limits_j G_j^v(\vec{c}_N)
             \\
             G_1^s(c_1)-G_1^s(c_N) & \dots & G_N^s(c_1)-G_N^s(c_N)
                 & \sum\limits_j G_j^v(\vec{c}_1)-\sum\limits_j G_j^v(\vec{c}_n)
     }
     \pmatrix{q_1 \\ \vdots \\ q_N \\ \gamma}
             = \pmatrix{-\hat{n}_1\cdot V_\infty(c_1) 
                 \\ \vdots \\ -\hat{n}_N\cdot V_\infty(c_N)
                 \\ V_\infty(c_N) - V_\infty(c_1)}
\end{align}

\begin{align}
    \Rightarrow &
        \mathbb{G}{\bf Q}
        = \mathbb{V}
    \\ \Rightarrow &
        {\bf Q}
        = \mathbb{G}^{-1}\mathbb{V}
\end{align}

Notice that the last row of matrix $\mathbb{G}$ as written above is a row of vectors which, if broken down into components, over-defines the system. Hence, instead of imposing the condition $\vec{V}(\vec{c}_1) = \vec{V}(\vec{c}_N)$, let's switch it to $\|\vec{V}(\vec{c}_1)\| = \|\vec{V}(\vec{c}_N)\|$ as to impose the same pressure between upper and lower surface having the same effect of connecting the streamlines at the trailing edge. Since the normal components of the panel-induced velocity cancels with the freestream, the only component on both trailing edge panels is tangential velocity, hence the condition now becomes

\begin{align}
    -\hat{t}_1\cdot\vec{V}(\vec{c}_1) = \hat{t}_N\cdot\vec{V}(\vec{c}_N)
\end{align}

\begin{align}
    -\hat{t}_1\cdot(\sum\limits_j q_j G_j^s(\vec{c}_1) + 
                        \gamma\sum\limits_j G_j^v(\vec{c}_1) + V_\infty(\vec{c}_1))
    = \hat{t}_N\cdot(\sum\limits_j q_j G_j^s(\vec{c}_N) + 
                        \gamma\sum\limits_j G_j^v(\vec{c}_N) + V_\infty(\vec{c}_N))
.\end{align}

(The negative sign in front of $\hat{t}_1$ is to point it up from $B$ to $A$ out leaving the trailing edge) 

The system of equations now is expressed as

\begin{align}
    \pmatrix{\hat{n}_1\cdot G_1^s(c_1) & \cdots & \hat{n}_1\cdot G_N^s(c_1)
                & \hat{n}_1\cdot \sum\limits_j G_j^v(\vec{c}_1)
             \\
             \vdots
             \\
             \hat{n}_N\cdot G_1^s(c_N) & \cdots & \hat{n}_N\cdot G_N^s(c_N)
                & \hat{n}_N\cdot \sum\limits_j G_j^v(\vec{c}_N)
             \\
             -\hat{t}_1\cdot G_1^s(c_1)-\hat{t}_N\cdot G_1^s(c_N) & \dots & 
                 -\hat{t}_1\cdot G_N^s(c_1)- \hat{t}_N\cdot G_N^s(c_N)
                 & -\hat{t}_1\cdot\sum\limits_j G_j^v(\vec{c}_1)-
                     \hat{t}_N\cdot \sum\limits_j G_j^v(\vec{c}_n)
     }
     \pmatrix{q_1 \\ \vdots \\ q_N \\ \gamma}
             = \pmatrix{-\hat{n}_1\cdot V_\infty(c_1) \\ 
                 \vdots \\ -\hat{n}_N\cdot V_\infty(c_N)
                 \\ \hat{t}_N\cdot V_\infty(c_N) + \hat{t}_1\cdot V_\infty(c_1)}
\end{align}

## Viscous Effects

### Laminar Boundary Layer

Properties of laminar boundary layers are calculated using Thwaites method as explained in Sec. 4-6.6 of White [1], and summarized as follows.

Lets define the non-dimensional parameter $\lambda = \frac{\theta^2}{\nu}\frac{\partial U_e}{\partial x}$, with $\theta$ momentum thickness, $\nu$ kinematic viscosity, $U_e$ $x$-velocity at the edge of boundary layer, and $x$ defined in the direction tangent to the surface of the body. It is found experimentally the following correlation:

\begin{align*}
    & \frac{U_e}{\nu}\frac{\partial \theta^2}{\partial x} \approx 0.45 - 6.0\lambda
    \\
    & \Rightarrow \frac{U_e}{\nu}\frac{\partial \theta^2}{\partial x} 
                \approx 0.45 - 6.0\frac{\theta^2}{\nu}\frac{\partial U_e}{\partial x} 
    \\
    & \Rightarrow \frac{1}{\nu}\Big( U_e\frac{\partial \theta^2}{\partial x} +
                                    6.0\theta^2\frac{\partial U_e}{\partial x} \Big)
                \approx 0.45 
    \\
    & \Rightarrow \frac{1}{\nu}\Big( U_e^6\frac{\partial \theta^2}{\partial x} +
                                    6.0U_e^5\theta^2\frac{\partial U_e}{\partial x} \Big)
                \approx 0.45 U_e^5
    \\
    & \Rightarrow \boxed{\frac{\text{d}}{\text{d}x}\Big( \theta^2U_e^6\Big)
                \approx  0.45 \nu U_e^5}
\end{align*}

This allows to solve for the growth of the momentum thickness along the surface given an initial condition $\theta_o$ at the leading edge. For a flat plate $\theta_o=0$, meanwhile for an airfoil $\theta_o = \sqrt{\frac{0.075\nu}{\frac{\partial U_e}{\partial x}(x=0)}}$. Once $\theta$ is solved numerically along the surface, the shape factor $H=\frac{\delta^*}{\theta}$ is calculated from the following empirical curve fit:

\begin{align}
    H(\lambda) \approx 2.0 + 4.14z -83.5z^2 + 854z^3 - 3337z^4 + 4576z^5
    , \quad z = 0.25-\lambda 
.\end{align}

Also, shear stress at the wall $\tau_w$ can be calculated by defining the non-dimensional parameter $S = \frac{\tau_w\theta}{\mu U_e}$, which follows the empirical relation

\begin{align}
    S(\lambda) \approx (\lambda + 0.09)^{0.62}
.\end{align}

It is important to note from White [1] p. 267, that "the accuracy of the Thwaites method is about $\pm5\%$ for favorable or mild adverse gradients but may be as much as $\pm15\%$ near the separation point." The separation point is encountered as $\lambda \rightarrow -0.09$.

All these calculation are implemented in the function `thwaites()`.

### Turbulent Boundary Layer

Properties of turbulent boundary layers are calculated using Head's method summarized as follows.

<img src="img/blayer00.png" alt="Pic here" style="width: 300px;"/>

Defining a control volume encompasing the boundary layer as shown above, $V_{ye}$ (the velocity that freestream is being entrained into the boundary layer) is calculated from conservation of mass as



\begin{align}
    V_{ye} = \int\limits_0^\delta \frac{\partial u}{\partial x} \text{d}y
,\end{align}

which develops in the following relation

\begin{align*}
        \frac{V_{ye}}{V_e} = & \frac{1}{V_e}\frac{\partial}{\partial x}
                                \int\limits_0^\delta  u \text{d}y
    \\
        = & \frac{1}{V_e}\frac{\partial}{\partial x}
                                \int\limits_0^\delta  (V_e - V_e + u) \text{d}y
    \\
        = & \frac{1}{V_e}\frac{\partial}{\partial x} V_e
            \int\limits_0^\delta  \big[ 1 - \big(1 - \frac{u}{V_e} \big) \big] \text{d}y
    \\
        = & \frac{1}{V_e}\frac{\partial}{\partial x} \big[ V_e (\delta - \delta^*) \big]
    \\
        = & \frac{1}{V_e}\frac{\partial}{\partial x} \big( V_e \theta H_1 \big)
            , \quad H_1 = \frac{\delta-\delta^*}{\theta}
.\end{align*}

It is assumed that $\frac{V_{ye}}{V_e}$ is a function only of the shape factor $H$, *i.e.,* $\frac{V_{ye}}{V_e} = f(H_1)$, and it is found empirically (Cebeci and Bradshaw) that

\begin{align}
    f(H_1) = 0.0306(H_1-3)^{-0.6169}
\end{align}

and 

\begin{align}
    H = \begin{cases}
               1.1 + 0.86(H_1-3.3)^{-0.777} \quad & \text{if } H_1\ge 5.3 \\
               0.6778 + 1.1538(H_1-3.3)^{-0.326} \quad & \sim
            \end{cases}
\end{align}

or 

\begin{align}
    H_1 = \begin{cases}
               0.8234(H-1.1)^{-1.287} + 3.3 \quad & \text{if } H\le 1.6 \\
               1.5501(H-0.6778)^{-3.064} + 3.3 \quad & \sim
            \end{cases}
\end{align}

This leads to the following expression

\begin{align}\boxed{
    \frac{\partial}{\partial x} \big( V_e \theta H_1 \big) = V_e f(H_1)
}\end{align}

\begin{align}
    \Rightarrow \frac{\partial V_e}{\partial x} \theta H_1 + 
                V_e \frac{\partial \theta}{\partial x}  H_1 + 
                V_e \theta \frac{\partial H_1}{\partial x} = V_e f(H_1)
,\end{align}

here $\frac{\partial V_e}{\partial x}$ is given by the inviscid solution of the panel method, meanwhile $\frac{\partial \theta}{\partial x}$ is calculated from the momentum-integral equation

\begin{align}
    \frac{\partial \theta}{\partial x} = \frac{c_f}{2} - 
            \frac{\theta}{V_e}(H+2)\frac{\partial V_e}{\partial x}
,\end{align}

with $c_f$ provided by the Ludwieg-Tillman skin-friction law

\begin{align}
    c_f = 0.246 \times 10^{-0.678H}Re_\theta^{-0.268}
.\end{align}

Putting it all together, we end up with the following system of ODEs:

\begin{align}
    \bullet \quad & 
        \frac{\partial H_1}{\partial x} = \frac{f(H_1)}{\theta} - 
            \frac{H_1}{V_e} \frac{\partial V_e}{\partial x} - 
            \frac{H_1}{\theta} \frac{\partial \theta}{\partial x}
    \\
    \bullet \quad &
        \frac{\partial \theta}{\partial x} = \frac{c_f}{2} - 
                \frac{\theta}{V_e}(H+2)\frac{\partial V_e}{\partial x}
,\end{align}

and $H(x_o)=1.4$ from the flat plate solution, and $\theta(x_o)$ is taken from the laminar solution.

These calculations are implemented in the function `head()`.

### Laminar-turbulent Transition

Michel's method is used to identify transition from laminar to turbulent flow along any surface. This method predicts that transition will occur anytime the following condition is true:

\begin{align*}
    \text{Re}_\theta \gt 1.174(1+\frac{22400}{\text{Re}_x})\text{Re}_x^{0.46}
,\end{align*}

with $\text{Re}_x = \frac{\rho V_e x}{\mu}$, and $\text{Re}_\theta = \frac{\rho V_e \theta}{\mu}$.

### Drag Calculation

The total viscous drag $C_D$ is calculated using Squire-Young's formula:

\begin{align*}
    C_D = \bigg[ 2\frac{\theta_{TE}}{c}\Big(
                    \frac{V_{eTE}}{V_\infty}\Big)^{\frac{H_{TE}+5}{2}}
            \bigg]_{\text{upper}} +
            \bigg[ 2\frac{\theta_{TE}}{c}\Big(
                    \frac{V_{eTE}}{V_\infty}\Big)^{\frac{H_{TE}+5}{2}}
            \bigg]_{\text{lower}}
.\end{align*}

The friction drag $C_{Df}$ is calculated integrating the skin friction coefficient, and the pressure drag is calculated by sustracting the free-stream projection of friction drag from the total viscous drag as $C_{Dp} = C_D-\hat{v}_\infty\cdot C_{Df}$.

* NOTE: Since the current method doesn't account for flow separation, instead of using $\theta$ at the trailing edge, the code will use the point where the flow detaches.
* NOTE: Since the friction drag $C_{Df}$ is an integration over the surface, it will have a component in the direction of lift, however, it should be small.

## Aerodynamic Forces

For a body with vortex panels, the function `calc_liftKJ()` calculates the lift coefficient using the Kutta-Joukowski theorem:

\begin{align*}
    C_l = \frac{L'}{\frac{1}{2}\rho V_\infty^2 c} = 
            \frac{- \rho V_\infty\Gamma}{\frac{1}{2}\rho V_\infty^2 c} =
            \frac{- 2\Gamma}{V_\infty c}
,\end{align*}

where $L'$ is the lift per unit span, $\Gamma$ is the integration of the vortex strength circulation density $\gamma$ along the body, $V_\infty$ is the reference freestream velocity, and $c$ is the reference chord length.  Alternatively, the lift can be calculated from the integration of pressure of the panel inviscid solution through the function `calc_pressureforce()`.

Drag is calculated from the viscous effects of the boundary layer as already explained through the function `calc_blayer()`.

Pitching moment is calculated by integrating the pressure distribution along the surface and adding each panel's contribution of moment to the total moment. This is implemented in the function `calc_moment()` (for a more accurate calculation, the wall friction should be added to the pressure distribution, but the current implementation ignore this effects since they are small).
